```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python

def draw_file_gauss(fname):
    f = open(fname, 'r')
    lines = [] # f.readlines()
    for line in f:
        items = line.split(' ')
        lines.append(items)
    
    mu = float(lines[0][0])
    std = float(lines[0][1])
    data = []
    for d in lines[1]:
        if d != '\n':
            data.append(float(d))
    
    print('mu: ', mu, 'std: ', std)
    # print('data: ', data)
    
    np_mu = np.mean(data)
    np_std = np.std(data)
    print('np_mu: ', np_mu, 'np_std: ', np_std)
    
    x = np.linspace(mu - 3*std, mu + 3*std, 100)

    plt.plot(x, stats.norm.pdf(x, mu, std))
    plt.hist(data, density=True, alpha=0.5, color='steelblue')
    
    # rv = stats.norm.pdf(x, mu, std)
    # y = norm.rvs(size=1000)*std + mu
    #plt.hist(y, density=True, histtype='stepfilled', alpha=0.5)
    plt.title(fname)
    plt.show()

def draw_file_gamma(fname):
    f = open(fname, 'r')
    lines = [] # f.readlines()
    for line in f:
        items = line.split(' ')
        lines.append(items)
    
    mu = float(lines[0][0])
    std = float(lines[0][1])
    data = []
    for d in lines[1]:
        if d != '\n':
            data.append(float(d))
    
    print('mu: ', mu, 'std: ', std)
    # print('data: ', data)
    
    np_mu = np.mean(data)
    np_std = np.std(data)
    print('np_mu: ', np_mu, 'np_std: ', np_std)
    
    # lamb = (mu*mu*mu)/(std*std)
    # x = np.linspace(1./mu - 3*std, 1./mu + 3*std, 100)
    # x = np.linspace(stats.invgauss.ppf(0.01, mu, scale=lamb), stats.invgauss.ppf(0.99, mu, scale=lamb), 100)
    # print('alpha: ',alpha, 'beta: ', beta, 'theta: ', theta)
    
    mu = 10*mu
    std = 100*std 
    # mean = alpha * beta, std^2 = alpha * beta^2 
    beta = std*std/mu
    alpha = mu / beta
    
    print('alpha: ', alpha, 'beta: ', beta)
    data = 10*data
    # x = np.linspace(mu - 3*std, mu + 3*std, 100)
    x = np.linspace(stats.gamma.ppf(0.01, mu, 1./beta), stats.gamma.ppf(0.99, mu, 1./beta), 1000)
    plt.plot(x, stats.gamma.pdf(x, mu, 1./beta))
    # plt.plot(x, stats.norm.pdf(x, mu, std))
    # plt.plot(x, stats.invgauss.pdf(x, mu, scale=lamb))
    plt.hist(data, density=True, alpha=0.5, color='steelblue')
    
    # rv = stats.norm.pdf(x, mu, std)
    # y = norm.rvs(size=1000)*std + mu
    #plt.hist(y, density=True, histtype='stepfilled', alpha=0.5)
    plt.title(fname)
    plt.show()
# mu = 0
# variance = 1
# sigma = math.sqrt(variance)
# x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
# plt.plot(x, stats.norm.pdf(x, mu, sigma))
# plt.show()
```
