```python
import sys 
import numpy as np 



f = open('../data/central_point_std.log','r')
angle_m = []
X = []
Y = []
for line in f:
    items = line.split('\t')
    r = int(items[0])
    c = int(items[1])
    n = int(items[2])
    
    x_l = []
    y_l = []
    for i in range(n):
        x_l.append(float(items[i+3]))
    for i in range(n):
        y_l.append(float(items[i+3+n]))
    
    X.append(x_l)
    Y.append(y_l)
    
    print('x_l: ', x_l)
    print('y_l: ', y_l)

```

```python
# plot 
import matplotlib.pyplot as plt  

plt.plot(X[0], Y[0],'gs-')
plt.show()
```

```python
# Fit a second order polynomial to each
line_fit = np.polyfit(X[0], Y[0], 2)

# Generate x and y values for plotting
plotx = np.linspace(0.75, 7.5, num=50)
ploty = line_fit[0]*plotx**2 + line_fit[1]*plotx + line_fit[2]

print('line_fit parameters: ',line_fit)

plt.plot(plotx, ploty, color='red')
plt.plot(X[0], Y[0],'gs--')
# plt.xlim(0, img.shape[1])
# plt.ylim(img.shape[0], 0)
plt.show()
```

```python
f = open('../data/grid_point_std.log','r')
angle_m = []
X = []
Y = []
R = []
C = []
for line in f:
    items = line.split('\t')
    r = int(items[0])
    c = int(items[1])
    n = int(items[2])
    
    x_l = []
    y_l = []
    for i in range(n):
        x_l.append(float(items[i+3]))
    for i in range(n):
        y_l.append(float(items[i+3+n]))
    
    X.append(x_l)
    Y.append(y_l)
    R.append(r)
    C.append(c)
    
    print('x_l: ', x_l)
    print('y_l: ', y_l)
```

```python

LINE_FIT = []

for (r,c,x,y) in zip(R,C,X,Y):
    line_fit = np.polyfit(x,y,2)
    LINE_FIT.append(line_fit)
    print('r: ', r, 'c: ', c, ' poly_params: ',line_fit)

```

```python

m = np.column_stack((R, C, LINE_FIT))

# print(m)
# save into file 
# out_file = open('')

# with open('poly_fit_grid_point.log', 'wb') as f:
np.savetxt('poly_fit_grid_point.log', m, fmt='%.7f')
```
